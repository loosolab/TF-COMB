# Plot TFBS in genome view

We will first plot the TFBS sites from ChIP-seq of GM12878 as loaded previously:

```python
import tfcomb
C = tfcomb.CombObj().from_pickle("../data/GM12878.pkl")
```

This object already contains TFBS to show:

```python
C.TFBS[:10]
```

Now, using _.plot_TFBS_, we can show the TFBS in a certain window:

```python
C.plot_TFBS(window_chrom = "chr4",
            window_start = 2441831,
            window_end = 2441885)
```

The function can also take _fasta_ and _bigwigs_ to show additional tracks below the TFBS:

```python
C.plot_TFBS(window_chrom = "chr4",
            window_start = 2441831,
            window_end = 2441885,
            fasta="../data/hg38_chr4.fa",
            bigwigs = ["../data/GM12878_corrected.bw",
                       "../data/GM12878_footprints.bw"])
```

## Plot TFBS from motifs

For TFBS with strand information, e.g. from motifs, _.plot_TFBS_ will show the direction of the TFBS as arrows:

```python
C = tfcomb.CombObj().from_pickle("../data/GM12878_motifs.pkl")
```

```python
C.plot_TFBS(window_chrom = "chr4",
            window_start = 41709878,
            window_end = 41710236,
            bigwigs = ["../data/GM12878_corrected.bw",
                       "../data/GM12878_footprints.bw"])
```
